### EXERCISE 5

Write a Python program to check if all values in a list are **different**

```python
# YOUR CODE HERE:
def check_list_uniqueness(test_list):
  unique = True
  for item in test_list:
    amount = test_list.count(item) #counts the count of each element
    if amount > 1 :
      unique = False
  return unique
unique_list = [1,2,3,4]
non_unique_list = [1,2,2,3]
print(check_list_uniqueness(unique_list))
print(check_list_uniqueness(non_unique_list))
```

Write a Python program calculate the product, multiplying all the numbers of a given tuple.

Original Tuple:
(4, 3, 2, 2, -1, 18)
Product: -864

Original Tuple:
(2, 4, 8, 8, 3, 2, 9)
Product: 27648


Original Tuple:
(1, "Two", 3, "Four", 5)

```python
# YOUR CODE HERE:
# using https://www.geeksforgeeks.org/python-convert-numeric-words-to-numbers/ for dict
word_dict = {
    'One': 1,
    'Two': 2,
    'Three': 3,
    'Four': 4,
    'Five': 5,
    'Six': 6,
    'Seven': 7,
    'Eight': 8,
    'Nine': 9,
    'Zero': 0
}

def product_from_tuple(hopefully_int_tuple):
  product = 1
  for number in hopefully_int_tuple:
    if word_dict.get(number) is None: # should be (if number is int) with no else
      product *= number
    else:
      product *= word_dict[number]
  return product

# not until I saw what was supposed to come out that you wanted just the ints
first_tuple = (4, 3, 2, 2, -1, 18)
second_tuple = (2, 4, 8, 8, 3, 2, 9)
third_tuple = (1, "Two", 3, "Four", 5)
print(f"Original Tuple:{first_tuple} Product:{product_from_tuple(first_tuple)}")
print(f"Original Tuple:{second_tuple} Product:{product_from_tuple(second_tuple)}")
print(f"Original Tuple:{third_tuple} Product:{product_from_tuple(third_tuple)}")
```

### EXERCISE 8

Write a program to:
1. Generate a random number (kept secret from the user) using random.randint(0, 32)
2. Prompt the user "Guess a number" and record their input.
3. Tell them if they were correct, too high, or too low.

BONUS POINTS: limit the number of guesses to 5.

```python
# YOUR CODE HERE:
import random
secret = random.randint(0, 32)
trys = 5
guess = int(input(f"Guess the secret number between 0 and 32, you have {trys} trys left: "))
while guess != secret and trys != 1:
  trys -= 1
  if guess > secret:
    print("Your guess is too high!")
  else:
    print("Your guess is too low!")
  guess = int(input(f"Guess the secret number between 0 and 32, you have {trys} trys left: "))

if guess == secret:
  trys -= 1
  print(f"The secret number was {secret}, you guessed it in {5-trys} trys!")
else:
  print(f"The secret number was {secret}, you ran out of trys!")

```
